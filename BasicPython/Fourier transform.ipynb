```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
x = np.linspace(0, 6*2*np.pi, 1000)
y = 0.4*np.sin(x) + 4* np.sin(2*x) + 3.6 * np.sin(6*x)
plt.plot(x,y)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
n = 21
x = np.linspace(0, np.pi*2, n)
xx = np.linspace(0, np.pi*2, n*100)
t = np.arange(n)
fig, axs = plt.subplots(5,2, figsize=(8,6), sharex=True, sharey=True)
for i, ax in enumerate(axs.T.flatten()):
    ax.plot(x, np.sin(i*x), '.-')
    ax.plot(xx, np.sin(i*xx), lw=0.5, color='r')
    ax.set_title('%d cycles'%i)
plt.tight_layout()
```

```python
d = np.random.randn(n)
d = d - d.mean() + 2

fftd = np.fft.fft(d)
plt.plot(np.abs(fftd), '.-')
```
